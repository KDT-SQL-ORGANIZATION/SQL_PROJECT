```python
DP2 = pd.read_excel('./kindred.xlsx',header=[0,1],index_col=0)
DP2
```

```python
DP2.columns
```

```python
DP2.columns.get_level_values(1)=='유치원수 (개)'
```

```python
DP2.columns.get_level_values(1)=='원아수 (명)'
```

```python
DPgardon=DP2.T[DP2.columns.get_level_values(1)=='유치원수 (개)'].T
DPgardon.columns=['2013','2014','2015','2016','2017','2018','2019','2020','2021','2022']
DPgardon
```

```python
fig, axs = plt.subplots(6,3,figsize=(10,14),layout='constrained')
for a,ax in zip(DPgardon.index,axs.flat):
    ax.plot(DPgardon.columns,DPgardon.loc[a],'oc-',label=a)
    ax.set_xlabel('연도')
    ax.set_title(a)
    ax.set_ylabel('개',rotation=0)
plt.suptitle('전국 유치원 수(2013~2022)',fontsize='xx-large')
plt.show()
```

```python
DPchild=DP2.T[DP2.columns.get_level_values(1)=='원아수 (명)'].T
DPchild.columns=['2013','2014','2015','2016','2017','2018','2019','2020','2021','2022']
DPchild
```

```python
fig, axs = plt.subplots(6,3,figsize=(10,14),layout='constrained')
for a,ax in zip(DPchild.index,axs.flat):
    ax.plot(DPchild.columns,DPchild.loc[a],'oy-',label=a)
    ax.set_xlabel('연도')
    ax.set_title(a)
    ax.set_ylabel('명',rotation=0)
plt.suptitle('전국 유치원생수(2013~2022)',fontsize='xx-large')
plt.show()
```
